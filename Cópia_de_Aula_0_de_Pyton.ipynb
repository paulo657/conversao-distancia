```python
df.rename(columns={
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}, inplace=True)

display(df.columns)
```

```python
df['senioridade'] = df['senioridade'].replace({
    'SE': 'Sênior',
    'MI': 'Pleno',
    'EN': 'Júnior',
    'EX': 'Executivo'
})

display(df['senioridade'].value_counts())
```

```python
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Período',
    'FL': 'Freelancer'
})

display(df['contrato'].value_counts())
```

```python
display(df["tamanho_empresa"].value_counts())
```

```python
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'M': 'Médio',
    'L': 'Grande',
    'S': 'Pequeno'
})

display(df['tamanho_empresa'].value_counts())
```

```python
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
})

display(df['remoto'].value_counts())
```

```python
#Criar um novo datraframe do zero
# fillna = preencher os valores nulos
import numpy as np
df_salarios = pd.DataFrame({
    'nome': ["Ana", "Bruno", "Carlos", "Daniele"],
    'salarios': [4000, np.nan, 5000, np.nan]
})
df_salarios['salario_media'] = df_salarios['salarios'].fillna(df_salarios['salarios'].mean().round(2))
df_salarios
```

```python
#gerar a mediana
import numpy as np
#criar df de teste
df_salarios = pd.DataFrame({
    'nome': ["Ana", "Bruno", "Carlos", "Daniele", "Val"],
    'salarios': [4000, np.nan, 5000, np.nan, 100000]
})
#calcula a média salarial e subistitui os nulos pela média e arredonda os valores
df_salarios['salario_media'] = df_salarios['salarios'].fillna(df_salarios['salarios'].mean().round(2))
'''calcula a mediana e suibistirtui os nulos pela mediana
'''
df_salarios['salario_mediana'] = df_salarios['salarios'].fillna(df_salarios['salarios'].median())
df_salarios
```

```python
#Criar um novo datraframe do zero
# fillna = preencher os valores nulos
#adiconando a Val
import numpy as np
df_salarios = pd.DataFrame({
    'nome': ["Ana", "Bruno", "Carlos", "Daniele", "Val"],
    'salarios': [4000, np.nan, 5000, np.nan, 100000]
})
df_salarios['salario_mediana'] = df_salarios['salarios'].fillna(df_salarios['salarios'].median())
df_salarios
```
